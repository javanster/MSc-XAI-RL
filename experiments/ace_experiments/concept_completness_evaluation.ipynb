## CONCEPT COMPLETENESS EVALUATION

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_k_means(df, env):
    plt.figure(figsize=(10, 6))

    for layer in df['layer'].unique():
        df_layer = df[df['layer'] == layer]
        df_layer = df_layer.sort_values(by='k')
        plt.plot(df_layer['k'], df_layer['completeness_score'],  label=f"Layer {layer}")

    max_idx = df['completeness_score'].idxmax()
    max_val = df.loc[max_idx, 'completeness_score']
    max_k = df.loc[max_idx, 'k']

    plt.scatter(max_k, max_val, s=100, c='red', label=f"Max Completeness Score = {max_val:.3f}, at k = {max_k}") # type: ignore


    plt.xlabel("k (Number of Concepts)")
    plt.ylabel("Completeness Score")
    plt.title(f"k-means completeness scores, DNN CCM, {env}")
    plt.legend(title="Layer")
    plt.grid(False)
    plt.show()
```

```python
def plot_fc_means(df, env):
    plt.figure(figsize=(10, 6))

    for layer in df['layer'].unique():
        df_layer = df[df['layer'] == layer]
        df_layer = df_layer.sort_values(by='c')
        plt.plot(df_layer['c'], df_layer['completeness_score'],  label=f"Layer {layer}")

    max_idx = df['completeness_score'].idxmax()
    max_val = df.loc[max_idx, 'completeness_score']
    max_c = df.loc[max_idx, 'c']

    plt.scatter(max_c, max_val, s=100, c='red', label=f"Max Completeness Score = {max_val:.3f}, at c = {max_c}") # type: ignore


    plt.xlabel("k (Number of Concepts)")
    plt.ylabel("Completeness Score")
    plt.title(f"Fuzzy c-means completeness scores, DNN CCM, {env}")
    plt.legend(title="Layer")
    plt.grid(False)
    plt.show()
```

```python
def plot_manual(df, env):
    plt.figure(figsize=(8, 6))
    plt.plot(df['layer'], df['completeness_score'], linestyle='-')

    max_idx = df['completeness_score'].idxmax()
    max_val = df.loc[max_idx, 'completeness_score']
    max_layer = df.loc[max_idx, 'layer']

    # Mark the maximum completeness score with a red dot.
    plt.scatter(max_layer, max_val, s=100, c='red', label=f"Max Completeness Score = {max_val:.3f}, at layer = {max_layer}")

    plt.xlabel("Layer")
    plt.ylabel("Completeness Score")
    plt.title(f"Manual concepts completeness scores, DNN CCM, {env}")
    plt.xticks(df['layer']) 
    plt.grid(False)
    plt.legend()
    plt.show()
```

```python
df = pd.read_csv("../../rl_ccm_data/completeness_scores/gem_collector/denim-sweep-56/ccm_nn/k_means_completeness_scores_all_q.csv")

plot_k_means(df, "Gem Collector")

df = pd.read_csv("../../rl_ccm_data/completeness_scores/gem_collector/denim-sweep-56/ccm_nn/fc_means_completeness_scores_all_q.csv")

plot_fc_means(df, "Gem Collector")

df = pd.read_csv("../../rl_ccm_data/completeness_scores/gem_collector/denim-sweep-56/ccm_nn/manual_completeness_scores_all_q.csv")

plot_manual(df, "Gem Collector")

```

```python
def plot_dt_completeness_scores(df, env, c_or_k: str, method: str):
    plt.figure(figsize=(10, 6))

    for target_type in df['target_type'].unique():
        df_target_type = df[df['target_type'] == target_type]
        df_target_type = df_target_type.sort_values(by=c_or_k)
        plt.plot(df_target_type['max_depth'], df_target_type['completeness_score'],  label=f"Target type {target_type}")

    max_idx = df['completeness_score'].idxmax()
    max_val = df.loc[max_idx, 'completeness_score']
    max_depth = df.loc[max_idx, 'max_depth']

    plt.scatter(max_depth, max_val, s=100, c='red', label=f"Max Completeness Score = {max_val:.3f}, at max depth = {max_depth}") # type: ignore


    plt.xlabel("Max depth")
    plt.ylabel("Completeness Score")
    plt.title(f"{method} completeness scores, Decision tree CCM, {env}")
    plt.legend(title="Target type")
    plt.xticks([3,4,5])
    plt.grid(False)
    plt.show()
```

```python
df = pd.read_csv("../../rl_ccm_data/completeness_scores/gem_collector/denim-sweep-56/ccm_dt/k_means_concepts_completeness_scores.csv")

plot_dt_completeness_scores(df, "Gem Collector", "k", "k-means")

df = pd.read_csv("../../rl_ccm_data/completeness_scores/gem_collector/denim-sweep-56/ccm_dt/fc_means_concepts_completeness_scores.csv")

plot_dt_completeness_scores(df, "Gem Collector", "c", "Fuzzy c-means")

df = pd.read_csv("../../rl_ccm_data/completeness_scores/gem_collector/denim-sweep-56/ccm_dt/manual_concepts_completeness_scores.csv")

plot_dt_completeness_scores(df, "Gem Collector", "c", "Manual concepts")
```
